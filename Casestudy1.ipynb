```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
sampath=pd.read_csv("energy.csv")
print(sampath)
state1=sampath[sampath["State"]=="Andhra"]
print(state1)
rcb=sampath[sampath["State"]=="Karnataka"]
print(rcb)
#line
fig,diag=plt.subplots(2)
diag[0].plot(state1["Date"],state1["Consumption"])
diag[1].plot(rcb["Date"],rcb["Consumption"])

state1=sampath[sampath["State"]=="Telengana"]
print(state1)
rcb=sampath[sampath["State"]=="Tirupati"]
print(rcb)
#bar
fig,diag=plt.subplots(2)
diag[0].bar(state1["Date"],state1["Consumption"],color='green')
diag[1].bar(rcb["Date"],rcb["Consumption"])


#pie
state2='Kerala'
state3="Delhi"
ycp=[state3,state2]

state1=sampath[sampath['State']=="Kerala"]["Consumption"].sum()
rcb=sampath[sampath['State']=="Delhi"]["Consumption"].sum()
js=[state1,rcb]
fig,diag=plt.subplots(2)
diag[0].pie(js,labels=ycp,autopct='%1.1f%%')
```
