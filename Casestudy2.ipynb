```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sea
it=pd.read_csv("heart.csv")
print(it)
#Scatterchart age cholestrol
sea.scatterplot(x="age",y="cholestrol",data=it,hue="target")
plt.title("HEART ATTACK")
plt.xlabel('age')
plt.ylabel('cholestrol')
plt.show()    
#boxplot gender bp
sea.boxplot(x="gender",y="bp",data=it)
plt.show()
#barchart
colors=['red','green']
sea.countplot(x='target',data=it,palette=colors)
plt.show()
#FACETGRID(Map)
imran=sea.FacetGrid(data=it,col='sugar')
imran.map(sea.scatterplot,'age','bp')
plt.show()
sea.pairplot(data=it)
plt.show()
```
